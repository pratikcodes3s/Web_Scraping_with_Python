```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
speech_url = "https://www.theguardian.com/theguardian/2007/may/01/greatspeeches"

page = requests.get(speech_url)
# page
soup = BeautifulSoup(page.text,'html')

soup
```

```python
jln_speech = soup.find_all('p')
```

```python
jln_speech
```

```python
jln_speech_combined = [p.text for p in jln_speech]
```

```python
jln_speech_combined
```

```python
type(jln_speech_combined)
```

```python
speech_string = ' '.join(jln_speech_combined)
speech_string
```

```python
speech_string = speech_string.replace('[Victory to India].','')
```

```python
speech_string
```

```python
import re
```

```python
speech_no_punct = re.sub(r'[^\w\r\s]','',speech_string)
print(speech_no_punct)
```

```python
speech_lower_case = speech_no_punct.lower()
```

```python
re.split(r'\s+',speech_lower_case)
```

```python
speech_break = re.split(r'\s+',speech_lower_case)
```

```python
df = pd.DataFrame(speech_break).value_counts()
```
